```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape = (28, 28, 1), name='Vision_Input'),
    tf.keras.layers.Conv2D(filters=32, kernel_size=3, name='Convolution_Layer_1'),
    tf.keras.layers.Conv2D(filters=64, strides=2, kernel_size=3, name='Convolution_Layer_2'),
    tf.keras.layers.Conv2D(filters=128, strides=2, kernel_size=3, name='Convolution_Layer_3'),
    tf.keras.layers.Dense(units=10, name='Vision_Output')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy()],
)
plot_model(model, to_file='vision_')
```
